```python
import requests, json
import pandas as pd
import numpy as np
import random
```

```python
tags = pd.read_csv("https://raw.githubusercontent.com/Dst017/PDS_Project/main/battletags.csv", header = None).iloc[0, :]
tags = tags.map(lambda x: x.replace("#", "-"))
tags = tags.to_numpy()
```

```python
indices = random.sample(range(tags.shape[0]), 1500)
tags = tags[indices]
```

```python
def wantedStats(json):
    all_dct = json["careerStats"]["allHeroes"]
    json["careerStats"] = all_dct
    return json

data = []
for i in range(tags.shape[0]):
    if (i % 100 == 0): print(i)
    try:
        call = "https://ow-api.com/v1/stats/pc/us/" + tags[i] + "/complete"
        response = requests.get(call)
        r_dict = response.json()
        if ("error" in r_dict):
            if (r_dict["error"] != "Player not found"):
                print(r_dict["error"], i, tags[i])
        elif (r_dict["private"]):
            continue
        elif (len(r_dict["competitiveStats"]["careerStats"]) == 0):
            continue
        else:
            data.append(wantedStats(response.json()["competitiveStats"]))
    except:
        print(i, tags[i])

with open('data.txt', 'w') as outfile:
    json.dump(data, outfile)
```
